```python
import pycaret
```

```python
from pycaret.regression import *
from pycaret.regression import RegressionExperiment
ExperimentoRegresion = RegressionExperiment()
import pandas as pd
```

```python
x_train=pd.read_csv('output/xtest.csv')
y_train=pd.read_csv('output/ytest.csv')
features=pd.read_csv('output/selected_features.csv')
x_train=x_train[features['0'].values]
```

```python
df=pd.concat([x_train, y_train], axis=1)
```

```python
ExperimentoRegresion.setup(df, target = 'price', session_id = 000)
```

```python
MejorModelo = ExperimentoRegresion.compare_models()
```

```python
ExperimentoRegresion.models()
```

Se puede ver que en todos los marcadores de rendimiento el mejor modelo es Extra Tree Regressor **(sklearn.ensemble._forest.ExtraTreesRegressor)** Luego lo sigue Ridge Bayesiano **(sklearn.linear_model._bayes.BayesianRidge)**, Random Forest Regressor **(sklearn.ensemble._forest.RandomForestRegressor)** y Hubber Regressor **(sklearn.linear_model._huber.HuberRegressor)**.
