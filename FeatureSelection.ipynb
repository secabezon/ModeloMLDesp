```python
# to handle datasets
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt

# to build the models
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
X_train = pd.read_csv('./output/xtrain.csv')
X_test = pd.read_csv('./output/xtest.csv')
y_train = pd.read_csv('./output/ytrain.csv')
y_test = pd.read_csv('./output/ytest.csv')
```

```python
sel_ = SelectFromModel(Lasso(alpha=0.001, random_state=0))

sel_.fit(X_train, y_train)
```

```python
sel_.get_support().sum()
```

```python
sel_ = SelectFromModel(Lasso(alpha=0.01, random_state=0))

sel_.fit(X_train, y_train)
```

```python
sel_.get_support().sum()
```

Se puede que ver que con un lambda de 0.001 no se elimina ninguna variable, es decir, que todas as variables tienen un grado de importancia en el modelo. Pero al hacer una analisis de sensibilidad y moviendo 0.01 se puede ver que se elimina una variable

```python

selected_feats = X_train.columns[(sel_.get_support())]

print('Número de Features en Total: {}'.format((X_train.shape[1])))
print('Número de Features Seleccionados: {}'.format(len(selected_feats)))
print('Características con coeficientes reducidos a cero: {}'.format(
    np.sum(sel_.estimator_.coef_ == 0)))
```

```python
# Imprimimos las Características o Features seleccionados
selected_feats
```

```python
for i in X_train.columns:
    if i not in selected_feats:
        print(i)
```
